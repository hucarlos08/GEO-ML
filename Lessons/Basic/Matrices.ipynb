# Matrix


## Identity and Inverse Matrices
An identity matrix is a matrix that does not change any vector when we multiply that vector by that matrix. Formally, $I_n\in\mathbb{R}^{n\times n}$, and
$$
\forall \mathbf x\in\mathbb R^n,\;\; I_n\mathbf{x} = \mathbf{x}.
$$

The matrix inverse of $A$ is denoted as $A^{−1}$, and it is defined as the matrix such that
$$
A^{−1}A = I_n.
$$

## Special kind of Matrices

1. **Symmetric matrix**:
$$A=A^T$$

1. **Orthogonal matrix**: square matrix whose rows are mutually orthonormal and whose columns are mutually orthonormal:
$$
A^TA = AA^T = I
$$

1. **Rectangular Diagonal Matrix**: A rectangular diagonal matrix is an $n\times d$ matrix in which each entry $(i, j)$ has a non-zero value if and only if $i = j$. For ejample
$$
\begin{bmatrix}1&0&0\\
0&4&0\\0&0&-3\\
0&0&0\end{bmatrix}
$$

1. **Upper and Lower Triangular Matrix**. A square matrix is an upper triangular matrix if all entries $(i, j)$ below its main diagonal (i.e., satisfying $i > j$) are zeros. The *lower* form is analogous.
$$
L = \begin{bmatrix}
\ell_{1,1} & 0 & 0 & \cdots & 0 \\
\ell_{2,1} & \ell_{2,2} & 0 & \cdots & 0 \\
\ell_{3,1} & \ell_{3,2} & \ell_{3,3} & \ddots & \vdots \\
\vdots & \vdots & \ddots & \ddots & 0 \\
\ell_{n,1} & \ell_{n,2} & \ldots & \ell_{n,n-1} & \ell_{n,n}
\end{bmatrix}
$$

## System of equations
We write down a system of linear equations:
$$
A\mathbf x = \mathbf b,
$$
where $A\in\mathbb R^{m\times n}$ is a known matrix, $\mathbf{b}\in\mathbb R^m$ is a known vector, and $\mathbf x\in\mathbb R^n$ isa vector of unknown variables we would like to solve for.

### Characteristics of the solutions
The are three cases for the linear system:
1. There is solution and is unique.
1. There is no solution
1. There is infinity solutions.

For **solve** the system for a matrix $A\in\mathbb{R}^{n\times n}$, and assuming that $A^{-1}$ exist:

\begin{align}
A^{-1}A\mathbf{x}&=&A^{-1}\mathbf b\
I_n\mathbf{x}&=&A^{-1}\mathbf b \
\mathbf{x}&=&A^{-1}\mathbf b
\end{align}
### Pseudoinverse
If $A\in\mathbb{R}^{m\times n}$ the only solution is given by the Moore-Penrose pseudoinverse. The pseudoinverse of $A$ is defined as a matrix:
$$
A^+ = \lim_{\alpha\to0}(A^TA+\alpha I_n)^{-1}A^T.
$$

1. When $A$ has **more columns than rows**, then solving a linear equation $A^+$ provides one of the many possible solutions. Specifically, it provides the solution $\mathbf x=A^+\mathbf b$ with minimal Euclidean norm $||\mathbf x||_2$
1. When $A$ has more rows than columns, it is possible for there to be no solution.
In this case, using the pseudoinverse gives us the $\mathbf x$ for which $A\mathbf x$ is as close as possible to $\mathbf b$ in terms of Euclidean norm $||A\mathbf x − \mathbf y||_2$.


## Eigendecomposition


One of the most widely used kinds of matrix decomposition is called eigen-decomposition, in which we decompose a matrix into a set of *eigenvectors* and *eigenvalues*. An eigenvector of a square matrix $A$ is a non-zero vector $\mathbf v$ such that multiplication by $A$ alters only the scale of $\mathbf v$:
$$
A\mathbf v = \lambda\mathbf v.
$$

If $\mathbf v$ is an eigenvector of $A$, then so is any rescaled vector $s\mathbf v$ for $s\in\mathbb R,\; s\neq 0$ For this reason, we usually only look for unit eigenvectors.

Specifically, every real symmetric matrix can be decomposed into an expression using only real-valued eigenvectors and eigenvalues:
$$
A = Q\Lambda Q^T
$$
where $Q$ is an orthogonal matrix composed of eigenvectors of $A$, and $\Lambda$ is a
diagonal matrix. The eigenvalue $\Lambda_{i,i}$ is associated with the eigenvector in column i of $Q$, denoted as $Q_{:,i}$. Because $Q$ is an orthogonal matrix, we can think of $A$ as scaling space by $\Lambda$ in direction $\mathbf v_i$.

The eigendecomposition of a matrix tells us **many useful** facts about the matrix.

1. The matrix is singular **if and only if** any of the eigenvalues are zero.
1. A matrix whose eigenvalues are all positive is called **Positive definite**: A matrix $A\in\mathbb{R}^{n\times n}$ is positive definite if $\forall \mathbf{x}\in\mathbb{R}^n\neq\mathbf{0},\;\;\mathbf{x}^TA\mathbf{x}>0$.
1. A matrix whose eigenvalues are all positive or zero-valued is called **positive semidefinite**: A matrix $A\in\mathbb{R}^{n\times n}$ is semi-positive definite $\forall \mathbf{x}\in\mathbb{R}^n\neq\mathbf{0},\;\;\mathbf{x}^TA\mathbf{x}\geq0$.
1. If all eigenvalues are negative, the matrix is negative definite


### Example
Suppose that a matrix $A$ has n linearly independent eigenvectors, $\{\mathbf v_1,\dots, \mathbf v_n\}$, with corresponding eigenvalues $\{λ_1,...,λ_n\}$. The eigendecomposition of A is then given by
$$
A=Vdiag(\lambda)V^{-1}
$$
We have seen that constructing matrices with specific eigenvalues and eigenvectors allows us to stretch space in desired directions.

```python
import matplotlib.pyplot as plt
from scipy import linalg

import numpy as np

v = np.array([[1,1],[1,-1]])
l = np.array([[1,0],[0,1]])

A = np.dot(v, np.dot(l,linalg.inv(v)))

Np = 100
t = np.linspace(0,2*np.pi, Np)
p = np.zeros((2,Np))

r= 1.0 # Radius of the circle
for i,phi in enumerate(t):
  p[0,i] = r*np.cos(phi)
  p[1,i] = r*np.sin(phi)

projected = np.dot(A,p)

fig, (axs1, axs2) = plt.subplots(1, 2, figsize=(10,10), dpi=100)
axs1.plot(projected[0,:],projected[1,:])
axs1.quiver(0,0,v[0,0],v[1,0],scale=3, label=r"$v_1$", color='r')
axs1.quiver(0,0,v[0,1],v[1,1],scale=3, label=r"$v_2$", color='b')
axs1.axis('equal')
axs1.legend(loc='lower right')

axs1.set_title('Before multiplication', fontsize=10)

# Change the eigenvalues
l2 = np.array([[2,0],[0,0.5]])
A2 = np.dot(v, np.dot(l2,linalg.inv(v)))
projected2 = np.dot(A2,p)

#New axis
v2 = np.dot(A2,v)

axs2.plot(projected2[0,:],projected2[1,:])
axs2.quiver(0,0,v2[0,0],v2[1,0],scale=4.5, label=r"$\lambda_1v1$", color='black')
axs2.quiver(0,0,v[0,0],v[1,0],scale=3, label=r"$v_1$", color='r')
axs2.quiver(0,0,v[0,1],v[1,1],scale=3, label=r"$v_2$", color='b')
axs2.quiver(0,0,v2[0,1],v2[1,1],scale=4.5, label=r"$\lambda_2v2$", color='black')
axs2.axis('equal')
plt.legend(loc='lower right')
axs2.set_title('After multiplication', fontsize=10)
```

## Singular Value Decomposition (SVD)

Singular Value Decomposition (SVD) is a fundamental matrix factorization technique used in various applications such as data compression, dimensionality reduction, and principal component analysis. It decomposes any $m \times n$ matrix $A$ into three matrices:

$$
A = UDV^T
$$

### Components
- **$U$**: an $m \times m$ orthogonal matrix, whose columns are the left singular vectors.
- **$D$**: an $m \times n$ diagonal matrix, with non-negative real numbers called singular values.
- **$V^T$**: the transpose of an $n \times n$ orthogonal matrix $V$, whose columns are the right singular vectors.

The singular values in $D$ are ordered in descending magnitude. 

### Properties and Definitions
- **Left Singular Vectors ($U$)**: Orthonormal eigenvectors of $AA^*$.
- **Right Singular Vectors ($V$)**: Orthonormal eigenvectors of $A^*A$.
- **Singular Values ($\sigma$)**: Non-zero values are the square roots of the non-zero eigenvalues of both $AA^*$ and $A^*A$.

### Vector Relationships
For a singular value $\sigma$, there exist unit vectors $\mathbf{u}$ and $\mathbf{v}$ such that:
$$
A\mathbf{v} = \sigma\mathbf{u} \quad \text{and} \quad A^*\mathbf{u} = \sigma\mathbf{v}
$$
Here, $\mathbf{u}$ and $\mathbf{v}$ are the corresponding left and right singular vectors.

### Interpretations and Applications
- **Transformation Insight**: $A$ transforms vectors from one space to another, revealing directions of maximal variance when applied to data.
- **Data Representation**:
  - **Left Singular Vectors**: Indicate maximal variance directions in the transformed space.
  - **Right Singular Vectors**: Show original data space directions that correspond to these maximal variance directions when transformed.
  
Each left and right singular vector pair is linked through their singular value, capturing the strength of the transformation in that direction.

### Practical Uses
- **Data Compression**: Retaining the largest singular values reduces data dimensions while preserving essential features.
- **Dimensionality Reduction**: Helps identify principal components, simplifying data without significant information loss.
- **Pseudoinverse Calculation**: Useful for solving systems of linear equations and least squares problems.
- **Optimal Matrix Approximation**: The truncated SVD offers the best approximation of a matrix under specific norms.


### Example of SVD

```python
from skimage import data
from skimage.color import rgb2gray
from numpy.linalg import svd
from skimage import img_as_ubyte,img_as_float
import numpy as np
import matplotlib.pyplot as plt

gray_images = {
        "cat":rgb2gray(img_as_float(data.chelsea())),
        "astro":rgb2gray(img_as_float(data.astronaut())),
        "camera":data.camera(),
        "coin": data.coins(),
        "coffee":rgb2gray(img_as_float(data.coffee()))
}


def compress_svd(image,k):
    """
    Perform svd decomposition and truncated (using k singular values/vectors) reconstruction
    returns
    --------
      reconstructed matrix reconst_matrix, array of singular values s
    """
    U,s,V = svd(image,full_matrices=False)
    reconst_matrix = np.dot(U[:,:k],np.dot(np.diag(s[:k]),V[:k,:]))

    return reconst_matrix,s


def compress_show_gray_images(img_name,k):
    """
     compresses gray scale images and display the reconstructed image.
     Also displays a plot of singular values
    """
    image=gray_images[img_name]
    print("Image shape: ", image.shape)
    original_shape = image.shape
    reconst_img,s = compress_svd(image,k)
    fig,axes = plt.subplots(1,3,figsize=(10,6), dpi=100)

    # plot eigenvalues
    axes[0].plot(s)

    #Plot original image
    axes[1].set_title("Original image")
    axes[1].imshow(image,cmap='gray')
    axes[1].axis('off')

    # Compressed image
    compression_ratio =100.0* (k*(original_shape[0] + original_shape[1])+k)/(original_shape[0]*original_shape[1])
    axes[2].set_title("compression ratio={:.2f}".format(compression_ratio)+"%")
    axes[2].imshow(reconst_img,cmap='gray')
    axes[2].axis('off')
    fig.tight_layout()
```

```python
compress_show_gray_images('coffee', 30)
```

## Eigen-decomposition vs SDV

```python
import numpy as np
import matplotlib.pyplot as plt

# Arbitrary matrix
A = np.array([[2, -1],
              [-1, 3]])

# Eigendecomposition
eigenvalues, eigenvectors = np.linalg.eig(A)

# Singular Value Decomposition
U, S, Vt = np.linalg.svd(A)

# Plot unit circle and its transformation under A
theta = np.linspace(0, 2*np.pi, 100)
circle = np.array([np.cos(theta), np.sin(theta)])

# Apply transformation by A
transformed_circle = A @ circle

# Plot the transformed unit circle
plt.figure(figsize=(8, 6))
plt.plot(circle[0], circle[1], label='Unit Circle', linestyle='--')
plt.plot(transformed_circle[0], transformed_circle[1], label='Transformed Circle', color='red')

# Plot eigenvectors
plt.quiver(0, 0, eigenvectors[0, 0], eigenvectors[1, 0], angles='xy', scale_units='xy', scale=1, color='blue', label='Eigenvector 1')
plt.quiver(0, 0, eigenvectors[0, 1], eigenvectors[1, 1], angles='xy', scale_units='xy', scale=1, color='green', label='Eigenvector 2')

# Plot left singular vectors
plt.quiver(0, 0, U[0, 0], U[1, 0], angles='xy', scale_units='xy', scale=1, color='purple', label='Left Singular Vector 1')
plt.quiver(0, 0, U[0, 1], U[1, 1], angles='xy', scale_units='xy', scale=1, color='orange', label='Left Singular Vector 2')

plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of Eigenvectors and Left Singular Vectors')
plt.legend()
plt.grid(True)
plt.show()
```

### Why the Difference?

1. **Matrix Properties**: Eigenvectors relate specifically to the properties of square matrices and especially to symmetric matrices where eigenvectors are orthogonal. In contrast, singular vectors are applicable in a broader context and are always orthogonal due to the properties of SVD.

2. **Directional Impact**: While eigenvectors tell us about the directions that are scaled by the matrix, singular vectors give us insight into how the structure of the data (or geometric shape, such as a circle turned into an ellipse) changes in both the original and transformed spaces. This includes understanding stretching and compression along principal axes, which are not necessarily aligned with the scaling directions implied by eigenvectors unless the matrix is symmetric.

3. **Orthogonality and Basis**: The left singular vectors form an orthogonal basis for the range of $A $, and the right singular vectors form an orthogonal basis for the domain of $A $. Eigenvectors do not necessarily form orthogonal bases unless the matrix is normal (i.e., $A^T A = A A^T $).
